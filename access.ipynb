```python
import pandas as pd
from data.database import get_db, ALL_TICKERS

db   = get_db()
data = db.get_period("test2")

def full_dataframe(data):
    """
    Convert the entire (T, N) numpy panel into one long-format
    DataFrame with columns: date, ticker, and all 7 features.
    """
    T = data["T"]
    N = data["N"]

    # Repeat dates for every stock, repeat tickers for every date
    dates_repeated   = np.repeat(data["dates"], N)          # length T*N
    tickers_tiled    = np.tile(ALL_TICKERS, T)               # length T*N

    df = pd.DataFrame({
        "date":         dates_repeated,
        "ticker":       tickers_tiled,
        "price":        data["prices"].flatten(),
        "high":         data["highs"].flatten(),
        "low":          data["lows"].flatten(),
        "volume":       data["volumes"].flatten(),
        "available":    data["availability"].flatten(),
        "return_1d":    data["return_1d"].flatten(),
        "return_5d":    data["return_5d"].flatten(),
        "return_20d":   data["return_20d"].flatten(),
        "return_60d":   data["return_60d"].flatten(),
        "vol_20d":      data["vol_20d"].flatten(),
        "cs_spread":    data["cs_spread"].flatten(),
        "amihud":       data["amihud"].flatten(),
    })

    return df

import numpy as np   # needed for np.repeat and np.tile above

df = full_dataframe(data)
print(df.shape)        # (T * 67, 14)
print(df.head(20))
```
